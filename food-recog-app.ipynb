```python
from fastai.vision.all import *
import gradio as gr
import timm
```

```python
im = PilImage.create('/kaggle/input/food-testing/007058.jpg')
im.thumbnail((224,224))
im
```

```python
# load the learner
model = 'foodmodel.pkl'
learn = load_learner(model)
```

```python
learn.predict(im)
```

```python
categories = learn.dls.vocab

def classify_image(img):
    pred,idx,probs = learn.predict(img)
    return dict(zip(categories, map(float,probs)))
```

```python
classify_image(im)
```

```python
#export
image = gr.inputs.Image(shape=(192, 192))
label = gr.outputs.Label()
examples = ['/kaggle/input/food-testing/007058.jpg']
```

```python
#export
intf = gr.Interface(fn=classify_image, inputs=image, outputs=label, examples=examples)
intf.launch(inline=False)
```

```python

def predict_food(image_name):
    # Open the image
    img = PIL.Image.open(image_name)

    # Resize the image to 224x224
    img.thumbnail((224, 224))

    # Convert the image to a tensor
    img_tensor = tensor(img).float()/255

    # Make a prediction using the loaded learner
    pred, pred_idx, probs = learn.predict(img_tensor)

    # Return the predicted class and the probabilities of all classes
    return {'prediction': str(pred), 'probabilities': {learn.dls.vocab[i]: round(float(probs[i]), 4) for i in range(len(probs))}}

# Define the inputs and outputs for the Gradio interface
inputs = gr.inputs.Image(label='Input Image', type='file')
outputs = gr.outputs.JSON(label='Predictions')

# Create the Gradio interface
gr.Interface(predict_food, inputs, outputs, title='Food Classification App', description='Classify images of different types of food').launch();
```
